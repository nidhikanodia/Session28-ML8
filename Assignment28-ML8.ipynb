```python
#Find the frequency of words in a webpage

#Import the required libraries
from bs4 import BeautifulSoup
import urllib.request
import nltk
import re
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

#Load the data
response = urllib.request.urlopen('http://php.net/')
html = response.read()
soup = BeautifulSoup(html,"html5lib")
```

```python
#Get the words from the webpage
webpage_words = soup.get_text()
```

```python
#Using the regular expressions, remove all the unwanted data from the text
#‘\w+’ will match arbitrary sequences of alphanumeric characters and underscores
tokens = re.findall('\w+', webpage_words)

#Print few sample tokens as sanity test
tokens[:10]
```

```python
#Convert all the uppercase letters to lowercase letters
#Initialize the new list
words = []

#Loop through list tokens and make lower case
for word in tokens:
    words.append(word.lower())
 
#Print few sample words as sanity test
words[:8]
```

```python
#Get the frequency distribution of words using Natural Language Tool kit (nltk)
counts = nltk.FreqDist(words)
```

```python
print("The frequency distribution of words is:")
print(counts.pformat(len(counts)))
```

```python
#Plot the frequency distribution of top 100 words
sns.set(rc={'figure.figsize':(20,10)})
sns.set_style('darkgrid')
counts.plot(100)
```
